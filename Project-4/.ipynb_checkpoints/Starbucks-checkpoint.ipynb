# Exploratory Data Analysis
## From Udacity: 
### Data Sets

The data is contained in three files:

* portfolio.json - containing offer ids and meta data about each offer (duration, type, etc.)
* profile.json - demographic data for each customer
* transcript.json - records for transactions, offers received, offers viewed, and offers completed

Here is the schema and explanation of each variable in the files:

**portfolio.json**
* id (string) - offer id
* offer_type (string) - type of offer ie BOGO, discount, informational
* difficulty (int) - minimum required spend to complete an offer
* reward (int) - reward given for completing an offer
* duration (int) - time for offer to be open, in days
* channels (list of strings)

**profile.json**
* age (int) - age of the customer 
* became_member_on (int) - date when customer created an app account
* gender (str) - gender of the customer (note some entries contain 'O' for other rather than M or F)
* id (str) - customer id
* income (float) - customer's income

**transcript.json**
* event (str) - record description (ie transaction, offer received, offer viewed, etc.)
* person (str) - customer id
* time (int) - time in hours since start of test. The data begins at time t=0
* value - (dict of strings) - either an offer id or transaction amount depending on the record


## 1)  portfolio.json

```python
portfolio.head()
```

From the information provided by Udacity, the portfolio.json file contains offer IDs and metadata on the offers. The schema is as follows:

**portfolio.json**
* id (string) - offer id
* offer_type (string) - type of offer ie BOGO, discount, informational
* difficulty (int) - minimum required spend to complete an offer
* reward (int) - reward given for completing an offer
* duration (int) - time for offer to be open, in days
* channels (list of strings)

We can see that the offer_type is categorical, as is channels. I can expand these into logical categorical columns.

```python
portfolio.shape
```

We can't use Pandas get_dummies to encode the channels variable due to the fact that the categories are contained in lists.

```python
channels = portfolio.channels
data = np.zeros((10,4))
for i in np.arange(0,portfolio.shape[0]):
        if 'web' in channels[i]:
            data[i,0]=1
        if 'email' in channels[i]:
            data[i,1]=1
        if 'mobile' in channels[i]:
            data[i,2]=1
        if 'social' in channels[i]:
            data[i,3]=1
df_channels = pd.DataFrame(data=data,
                           columns = ['web','email','mobile','social'])
df_channels = df_channels.astype(int)
df_channels
```

```python
df_offer_type = pd.get_dummies(portfolio.offer_type)
```

I will drop the channels column, but will keep the offer_type channel for grouping purposes.

```python
df_portfolio = pd.concat([portfolio,df_channels,df_offer_type],axis=1)
df_portfolio.drop(['channels'],axis=1,inplace=True)
df_portfolio
```

Let's change id to offer_id so we don't confuse it with customer ID.

```python
df_portfolio = df_portfolio.rename(columns={'id':'offer_id'})
df_portfolio
```

Here we have the new portfolio data set with the channels and offer_type columns expanded into logical categorical columns.

This data set is small enough that we can visually confirm that there is no missing or anomalous data.

# Modeling

## Model selection
Based on this handy chart from sci-kit learn, I have chosen to focus first on SGD (Stochastic gradient Descent). This classifier is sensitive to feature scaling, so I will create a pipeline to transform the data, create the model, and fit the data to the model. SGD is also sensative to hyperparameter tuning, so I will include a cross-validation gridsearch to tune the model.

Based on the scaler documentation below, I've chosen StandardScaler to do the preprocessing transformation. This removes the mean and scales the data to unit variance -- this is called for in the documentation for the SGD classifier. This scaler is sensative to outliers, but from our exploratory analysis of the data, we don't have any large outliers in our non-binary data.

    Due to the fact that my response variable is binary (I'm looking to predict if an offer will be completed), I will set the SGD classifier to 'log-loss', which mimics a logistic regression over many classifier variables.
Just kidding! Log-loss isn't available in the version of scikit-learn that is supports my current python build. So, I'll go ahead and run both the SGD classifier with it's default loss function, and run a logistic regression pipeline to compare the two models.

https://scikit-learn.org/stable/machine_learning_map.html

https://scikit-learn.org/stable/modules/sgd.html

https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#sphx-glr-auto-examples-preprocessing-plot-all-scaling-py

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),   # Step 1: Standard scaler
    ('sgd_classifier', SGDClassifier(fit_intercept=False,   # Data is already centered
                                     random_state=42))  # For repeatability
])

pipeline.fit(X_train,y_train)

model_params = pipeline.get_params()
print(f'Best parameters: {model_params}')

y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
param_grid = {
    'sgd_classifier__alpha': [1e-4, 1e-3, 1e-2],  # Regularization strength
    'sgd_classifier__max_iter': [1000, 2000]  # Number of iterations
}

grid_search = GridSearchCV(pipeline, param_grid, scoring='accuracy')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f'Best parameters: {best_params}')

y_pred = grid_search.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Create the logistic regression pipeline
pipeline_log = Pipeline([
    ('scaler', StandardScaler()),   # Step 1: Standard scaler
    ('logistic_regression', LogisticRegression(solver='saga', # More robust for larger datasets
                                               random_state=42))  # For repeatability
])

pipeline_log.fit(X_train,y_train)

model_params = pipeline_log.get_params()
print(f'Best parameters: {model_params}')

y_pred = pipeline_log.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
param_grid = {
    'logistic_regression__C': [0.1, 1, 5],  # Inverse of regularization strength
    'logistic_regression__penalty': ['l1', 'l2'],  # Regularization type
    'logistic_regression__tol': [1e-3, 1e-4]  # Tolerance for stopping criteria
}

grid_search_log = GridSearchCV(pipeline_log, param_grid, scoring='accuracy')

grid_search_log.fit(X_train, y_train)

best_params = grid_search_log.best_params_
print(f'Best parameters: {best_params}')

y_pred = grid_search_log.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Create the logistic regression pipeline
pipeline_log = Pipeline([
    ('scaler', StandardScaler()),   # Step 1: Standard scaler
    ('logistic_regression', LogisticRegression(solver='saga', # More robust for larger datasets
                                               C= 0.1,
                                               max_iter= 100,
                                               penalty= 'l1',
                                               tol= 0.0001,
                                               random_state=42))  # For repeatability
])

pipeline_log.fit(X_train,y_train)

model_params = pipeline_log.get_params()
print(f'Best parameters: {model_params}')

y_pred = pipeline_log.predict(X_test)

print(classification_report(y_test, y_pred))
```
